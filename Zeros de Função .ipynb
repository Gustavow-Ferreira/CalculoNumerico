```python
# Ex.: Função x³- 9x + 5 em intervalo I = [-5,5], passo = 1 e precisão = 0,01
# Dados iniciais
e = 0.01
a = 0.5
b = 4
h = 0.5
x = []
y = []
raiz_a = []
raiz_b = []
fix = 4 

def f(x):
    return round((x**3)-(9*x)+ 3,fix) # Questão da prova
    #return (x**3)-(9*x)+5 # Sala de aula
    #return (x**3)+4*(x**2)-10 # Questão 24
    #return (x**3)+(x**2)-(10*x)-8 # Questão 23
def dydx(x):
    return (3*(x**2))-9
    
# Isolamento das raízes
def tabela(a,b,p):
  print("Tabelamento\n-------------")
  print("x  | y")
  while a <= b:
    x.append(a)
    y.append(f(a))
    print(a, end=" | ")
    print(f(a))
    a += p
  for indice in range(0,len(y), 1):

    if indice+1 == len(y):
      break
    if (y[indice]*y[indice+1] < 0):
      raiz_a.append(x[indice])
      raiz_b.append(x[indice+1])

  print("\n")

def bisseccao(a, b):
    return (a+b)/2

def falsa_posicao(a, b):
    return (a*f(b)-b*f(a))/(f(b)-f(a))

def ponto_fixo(x): #phi(x)
    return (x**3+3)/9 

def newton(x):
    return x - f(x)/dydx(x)

def secante(a, b):
    return (a*f(b)-b*f(a))/(f(b)-f(a))

def metodos(m, a, b, x):
    if m == '1':
        return bisseccao(a, b)
    if m == '2':
        return falsa_posicao(a, b)
    if m == '3':
        return ponto_fixo(x)
    if m == '4':
        return newton(x)
    if m == '5':  
        return secante(a, b)
        
# Tratamento de raízes
def iteracoes(a, b, stop, m):
  
  if f(a)*f(b) < 0: # Verifica se há existência de raiz no intervalo
    k = 0
    if m == '1':
        print("Método da Bissecção\n")
        x = bisseccao(a, b)
    if m == '2':    
        print("Método da Falsa Posição\n")
        x = falsa_posicao(a, b)
    if m == '3':
        print("Método do Ponto Fixo\n")
        x = bisseccao(a, b) # Chute inicial no meio do intervalo
    if m == '4':
        print("Método de Newton-Raphson\n")
        x = bisseccao(a, b) # Chute inicial no meio do intervalo
    if m == '5':  
        print("Método da Secante\n")
        x = secante(a, b)
        
    if (abs(f(x)) < e):
        print("\nEncontrou a raiz")
        print("f(x)=",f(x))
    else:
        print("Valor de x",k, " é ",round(x,fix)) # Impressão de x e k
        k += 1
        while ((abs(f(x)) > e) and (stop > k)): # Enquanto módulo de f(x) maior que a precisão faça
          # Atualização de dados
          
          a, b = atualizar(m, a, b, x)
          x = round(metodos(m, a, b, x), fix)
          print("Valor de x",k, " é ",x) # Impressão de x e k
          k += 1
        
        print("f(x)=",f(x))
        
        if abs(f(x)) > e:  # Verifica se achou a raiz
          print("\nNão encontrou a raiz")
        else:
          print("\nEncontrou a raiz")
  else: # Se preservar sinal então
    print("Sem raiz entre o intervalo",a, "e ",b)
  if m != 5:
      print("Número de iterações(k):", k-1) # Impressão de iterações
  else: 
      print("Número de iterações(k):", k)
  print("\n")

def atualizar(m, a, b, x): # Usado apenas nos métodos de bissecção e falsa posição
    
    if (m == '1' or m == '2'):
        if f(a)*f(x) < 0: # Se não preservar sinal está entre [a,b]
            b = x
        else: # Se preservar então está entre [x,b]
            a = x
    if m == '5':
        a = b
        b = x
        
    return a, b



def set_parada():
  for indice in range(0, len(raiz_a)):

    print("\nHá raiz no intervalo: [", raiz_a[indice], ",", raiz_b[indice], "]\n")
    print("\nQuantas iterações?\n")
    k= int(input())
    print("")
    for m in range(1,6):   
        iteracoes(raiz_a[indice], raiz_b[indice], k+1, str(m))
    
def chamar_metodo():

    print("Continuar refinamento?\n 1.Sim\n 2.Não\n")
    opcao = input()
    while opcao != '2':

      if opcao == '1':
        set_parada()
        break
      else:
        limpar_terminal()
        tabela(a,b,h)
        chamar_metodo()
      break


tabela(a,b,h)
chamar_metodo()
```
