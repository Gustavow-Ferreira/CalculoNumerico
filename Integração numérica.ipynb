# Integração numérica
## Anotação: Aparentemente parece funcionar bem, mas em alguns casos com o arredondamento(round) é capaz de não realizar parte do tabelamento do intervalo (a,b) corretamente. 

```python
import sympy as sp

x = sp.symbols('x')
a = 1 # limite inferior
b = 3 # limite superior
y = [] # lista para valores de f(x)
fix = 4 # Fix casas decimais
print(f"Fix {fix}\n")
#funcao = sp.exp(x) # e^x
funcao = x * sp.log(x, sp.E) # x ln x
#funcao = 1/(x**2) # 1/x²
integral_indefinida = sp.integrate(funcao, x)
integral_definida = sp.integrate(funcao, (x, a, b))

print("Insira o número de partições: ", end="")
n = int(input())
if n < 0:
    n = 1
h = (b-a)/n

def f(t):
    return round(funcao.subs(x, t),fix)

def tabela(a, b):
    print("\n• Tabelamento\n-------------")
    print("x  | y")
    while a <= b:

        y.append(f(a)) # Armazenar os valores da função
        print(a, end=" | ")
        print(f(a))
        a += h
        a = round(a,fix)
        print("\n")


# Soma de Riemann a esquerda
def soma_esq(soma):
    
    for i in range(n): # Começa de i = 0 até n-1
        soma += y[i]
    soma *= h
    print(f"• Soma de Riemann à esquerda {round(soma, fix)} \n")
    return soma

# Soma de Riemann a direita
def soma_dir(soma):
    for i in range(1, n+1): # Começa de i = 1 até n
        soma += y[i]
    soma *= h
    print(f"• Soma de Riemann à direita {round(soma, fix)} \n")
    return soma

# Regra dos trapézios
def trapezio(soma):
    
    for i in range(1, n):
        soma += y[i]
    soma *= 2
    soma += y[0] + y[-1] # Soma os valores y0 e yn após iterações 
    soma *= h/2
    print(f"• Regra do trapézio {round(soma, fix)} \n")
    return soma

def terco_simpson(soma):
    
    if n % 2 == 0 and n >= 2:
        for i in range(1, n):
            if i % 2 == 1: # Caso seja ímpar 
                soma += 4*(y[i])
            else: # Caso não seja ímpar 
                soma += 2*(y[i])
       
    soma += y[0] + y[-1] # Soma os valores y0 e yn após iterações 
    soma *= h/3
    print(f"• Regra do 1/3 de simpson {round(soma,fix)} \n")
    return soma

# Cálculo dos erros absoluto e relativo
def calc_erro(soma):
    
    Eabs = abs(soma - round(integral_definida.evalf(), fix))
    Erel = Eabs/abs(round(integral_definida.evalf(), fix))

    print(f"Erro absoluto é {round(Eabs,fix)}") # Arredondando para o erro absoluto com fix casas
    print(f"Erro relativo é {round(Erel*100, fix)} % \n")
    # math.trunc(Erel*100) Truncamento para percentual

tabela(a,b)
print("Integral Indefinida:", integral_indefinida, "\n")
print("Integral Definida:", integral_definida, "\n")

# Vai realizar o cálculo e impressões dos erros de cada soma 
calc_erro(soma_esq(0))
calc_erro(soma_dir(0))
calc_erro(trapezio(0))
calc_erro(terco_simpson(0))
```
